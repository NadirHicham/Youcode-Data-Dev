```python
import geopandas as gpd
import folium
from folium.plugins import AntPath
from geopy.distance import distance
import ipywidgets as widgets
from IPython.display import clear_output

m = folium.Map(location=[32.24556456855282,-8.532570502180334],zoom_start=15,)

df_places = gpd.read_file('data/youssoufia.json')


def getMap(fromPos,toPos):

    clear_output()

    fromPosition = df_places.loc[df_places['title'] == fromPos ]['geometry']
    toPosition = df_places.loc[df_places['title'] == toPos ]['geometry']
    
    for _,p in df_places.iterrows():
        marker = folium.Marker(
            location=[p['geometry'].y,p['geometry'].x],
            popup=p['title'],
            icon=folium.Icon(icon="home"),
        )
        marker.add_to(m)
    
    if float(fromPosition.x) != float(toPosition.x):
        pos1 = (float(fromPosition.y),float(fromPosition.x))
        pos2 = (float(toPosition.y),float(toPosition.x))
        coords = (pos1,pos2)
        
        AntPath(coords,color='blue',dash_array=[30,15],tooltip=f'{round(distance(pos1,pos2).km,2)} KM',weight=5, opacity=.85).add_to(m)

    display(w1,w2,btn,m)


titleFromPosition = df_places['title'][0]
titletoPosition = df_places['title'][1]

w1 = widgets.Dropdown(
    options=df_places['title'],
    value=titleFromPosition,
    description='From:',
    disabled=False,
)

w2 =  widgets.Dropdown(
    options=df_places['title'],
    value=titletoPosition,
    description='To:',
    disabled=False,
)

btn = widgets.Button(
    value=False,
    description='Calcule',
    disabled=False,
    button_style='success',
    tooltip='Description',
    icon='check'
)

def getFirstPosition(title):
    global titleFromPosition
    titleFromPosition = title

def getSecondPosition(title):
    global titletoPosition
    titletoPosition = title

w1.observe(lambda x:getFirstPosition(x.new),'value')
w2.observe(lambda x:getSecondPosition(x.new),'value')

btn.on_click(lambda x:getMap(titleFromPosition,titletoPosition))

display(w1,w2,btn)
```
